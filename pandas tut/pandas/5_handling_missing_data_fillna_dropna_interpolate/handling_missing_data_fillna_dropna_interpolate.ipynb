## <font color="maroon"><h4 align="center">Handling Missing Data - fillna, interpolate, dropna</font>

```python
import pandas as pd
df = pd.read_csv("weather_data.csv",parse_dates=['day'])
type(df.day[0]) # coloum string day => date day coloum
df
```

```python
df.set_index('day',inplace=True)
df
```

## <font color="blue">fillna</font>

<font color="purple">**Fill all NaN with one specific value**</font>

```python
new_df = df.fillna(0) # fillna( pass value )
new_df
```

<font color="purple">**Fill na using column names and dict**</font>

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'No Event'
})  # fillna( dict )
new_df
```

<font color="purple">**Use method to determine how to fill na values**</font>

```python
new_df = df.fillna(method="ffill")
new_df # carry forward previous day value
```

```python
new_df = df.fillna(method="bfill")
new_df # carry forward next day value
```

<font color="purple">**Use of axis**</font>

```python
new_df = df.fillna(method="bfill", axis='columns') 
new_df # axis is either "index" or "columns"
```

<font color="purple">**limit parameter**</font>

```python
new_df = df.fillna(method="ffill",limit=1)
new_df
```

### <font color="blue">interpolate</font>

```python
new_df = df.interpolate()
new_df # divide prev and next into no of middle value parts
```

```python
new_df = df.interpolate(method="time") 
new_df # guess val according to time
```

**Notice that in above temperature on 2017-01-04 is 29 instead of 30 (in plain linear interpolate)**

**There are many other methods for interpolation such as quadratic, piecewise_polynomial, cubic etc. 
Just google "dataframe interpolate" to see complete documentation**

### <font color="blue">dropna</font>

```python
new_df = df.dropna()
new_df # drop => all row conataining n.a
```

```python
new_df = df.dropna(how='all')
new_df # keep => if row having some data
```

```python
new_df = df.dropna(thresh=1)
new_df # keep => at least one valid value 
```

### <font color="blue">Inserting Missing Dates</font>

```python
dt = pd.date_range("01-01-2017","01-11-2017") # create date range
idx = pd.DatetimeIndex(dt)  # create data time index
df.reindex(idx) # Inserting Missing Dates
```
